```python
from graphviz import Digraph

def create_executive_summary_visual():
    # Create a directed graph
    dot = Digraph(comment='From Forecasts to Decisions', format='png')
    
    # Set graph attributes for a clean, modern look
    dot.attr(rankdir='LR',  # Left to Right flow
             splines='ortho', 
             nodesep='0.6', 
             ranksep='0.8',
             fontname='Arial')

    # Define Node Styles
    dot.attr('node', shape='box', style='rounded,filled', fontname='Arial', fontsize='12')

    # --- THE OLD WAY (Problem) ---
    with dot.subgraph(name='cluster_problem') as c:
        c.attr(label='The "Technical" Trap (Problem)', color='#ffcccc', style='filled', fontcolor='#990000')
        c.node('Input', 'Raw Sales Data', fillcolor='white', color='grey')
        c.node('Model', 'Complex Black Box Model', fillcolor='white', color='grey')
        c.node('Metric', 'Minimize RMSE (Error)', fillcolor='#ffe6e6', color='red')
        c.node('Outcome', 'Ignored by Business\n(No Trust)', shape='parallelogram', fillcolor='#ffcccc', color='red')
        
        c.edge('Input', 'Model')
        c.edge('Model', 'Metric')
        c.edge('Metric', 'Outcome')

    # --- THE NEW WAY (Solution) ---
    with dot.subgraph(name='cluster_solution') as c:
        c.attr(label='Our Decision-Aware Framework (Solution)', color='#ccffcc', style='filled', fontcolor='#006600')

        # Key Nodes reflecting your "What I Built" section
        c.node('Data', 'Cleaned Data + Features', fillcolor='white')
        
        # Hybrid Logic
        c.node('Hybrid', 'Hybrid Model\n(ETS Trend + XGBoost Calendar)', fillcolor='white', color='blue')
        
        # The Decision Layer
        c.node('Context', 'Risk & Uncertainty\n(Prediction Intervals)', fillcolor='#e6ffe6', color='green')
        
        # Human Loop
        c.node('Action', 'Actionable Plan\n(Human-in-the-Loop)', shape='parallelogram', fillcolor='#ccffcc', color='green')

        c.edge('Data', 'Hybrid')
        c.edge('Hybrid', 'Context', label='Interpretable\nOutputs')
        c.edge('Context', 'Action', label='Decision\nSupport')

    # Connecting the shift
    dot.edge('Outcome', 'Data', style='invis') # Invisible edge to force layout

    # Render
    dot.render('forecasting_framework_visual', view=True)
    print("Visual generated: forecasting_framework_visual.png")

# Run generation
try:
    create_executive_summary_visual()
except Exception as e:
    print(f"Graphviz not installed or path not found. Error: {e}")
    print("To install: pip install graphviz (and install Graphviz software on OS)")
```
